## Mental Health Symptom Search 

   The months of March and April signaled the beginning of the pandemic, with no end in sight. The cases and death count in the United States were rising quickly and the bleak news translated into affecting not only people's physical health, but their mental health as well. As case and death count rose, Google Trends data  was collected on all symptoms that Americans were searching for including their mental and physical health. After narrowing down the mental health search symtoms, the cumulative searches showed a stronger correlation between COVID-19 case and death peaks and increased searches for mental health symptoms such as 'anxiety' and 'depression' during the early months of the pandemic. 
   
   These heat maps that were created by analyzing data from Google Trends to show the correlation between case and death rates in states and the searches for the symptom 'anxiety.' The deeper red colored squares on the heat map represents a stronger correlation, the deeper blue colored squares, the less strong of a correlation. The heat maps focus on one month and rank the states from greatest to least strongest correlation. The heat map for these correlations is shown here: 
  
#### __New cases and new deaths of COVID-19 and how they correlate to mental health symptom 'anxiety' Google search__
* In the month of March, states that had the highest mean case and death counts are colored red which means that there is a stronger correlation to searching for mental the health symptom of 'anxiety.'

#### __New deaths per month of COVID-19 and how it correlates to mental health symptom 'depression' Google search__
* In April, states that had the highest correlation between mean death counts and the mental health Google Trend search for 'depression' are more red and listed at the top of the heat map. These red colored states on the heat map suggest that there were more of an effect on the mental health of the residents of the states that had high death tolls.

![](graphs/newdeaths_depression.png)

Although there was not as strong of a correlation between the symptom search 'depression' and new cases of COVID-19 in the early months of the pandemic, the search for 'depression' also gained popularity when new cases of COVID-19 began increasing in the early summer months of May and June:
* In June, the search for 'depression' was still popular, despite it being the start of the summer. This heat map suggests that these mental health symptom searches expanded past the winter, a time when seasonal depression is common. This heat map suggests that the searches for mental health related symptoms were not due solely to seasonal depression but rather the ongoing pandemic.

  These heat maps support the hypothesis that during the months when cases and deaths were spiking, people were searching for mental health symptoms as well. This effect on the mental health of American citizens can be hypothesized to be due in part to depressing news, isolation in homes, and state of the world due to COVID-19. 

## Physical Health Symptom Search
   As the year moved into the earlier summer months of May and June, there was yet another spike in COVID-19 cases and deaths. As shown before, the mental health of Americans was being affected, even during the summer. The symptom search for 'depression' was still popular in early summer months, but the increase in COVID-19 related symptoms such as 'cough' and 'fever' were gaining popularity.

## Take aways
   It becomes apparent that during the earlier months of the pandemic, citizens of the U.S. were heavily searching for mental health related symptoms such as 'anxiety' and 'depression' which could be due in part to witnessing friends and loved ones being effected by the virus as well as the depressing news surrounding the ever increasing rise in numbers of the COVID-19 cases and deaths. The increase in mental health search queries in the winter could potentially be related to seasonal depression but it is more likely that the searches for mental health symptoms were related to the constant devastating COVID-19 news, which can be supported by the finding that 'depression' was still a popular search even in the early summer months when cases and deaths were rising exponentially. 
   
   Overall, there were many events occurring in the year of 2020 including Black Lives Matter protests in June and July, a national pandemic, and a presidential election in November. All of these could have had an effect on the mental and physical health of American citizens, but it is likely that the pandemic exacerbated these effects. 
   
   These Google Trend search queries provide a snapshot into the year that will forever be defined by the COVID-19 pandemic, supporting the relationship between how citizens are feeling both physically and mentally and the rise in cases and deaths caused by the virus. Although in the earlier months of the lockdowns it appears as though people were more often searching for mental health symptoms, when the second wave of the pandemic hit in May and June, people began searching for more physical health related symptoms. Overall, people were struggling both with their mental and physical health, but the increases in cases and deaths over time provides an interesting timeline of how people were feeling as the year progressed.
